```python
import matplotlib.pyplot as plt
import numpy as np
y = np.array([0, 2, 10, 5])
plt.plot(y,'o:r')
y = np.array([9, 3, 0, 4])
plt.plot(y, 'd--m')
y = np.array([4, 5, 13, 6])
plt.plot(y,'*-k')
y = np.array([8, 7, 1, 3])
plt.plot(y, 'h-.c')
plt.show()
y = np.array([0, 2, 10, 5])
plt.plot(y, marker = 'o', ms = 20, mec = 'r')
y = np.array([9, 3, 0, 4])
plt.plot(y, marker = 'd', ms = 10, mec = 'm')
y = np.array([4, 5, 13, 6])
plt.plot(y,marker = '*', ms = 40, mec = 'k')
y = np.array([8, 7, 1, 3])
plt.plot(y, marker = 'h', ms = 100, mec = 'c')
plt.show()
```
